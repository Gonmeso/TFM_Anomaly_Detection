```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
%matplotlib inline
```

## Create random data and plot it

```python
data = make_blobs(n_samples=[300,200], n_features=2, random_state=1, cluster_std=1.8, )
fig, ax = plt.subplots(ncols=2, figsize=(10,5))
ax[0].scatter(x=data[0][:,0], y=data[0][:,1])
ax[1].scatter(x=data[0][:,0], y=data[0][:,1], c = data[1])
plt.savefig('clustering_example.png')
```
