## Create Data

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib inline
```

## Plot the data without PCA

```python
data = make_blobs(100, 2, cluster_std=1, random_state=101)

plt.scatter(x=data[0][:,0], y =data[0][:,1])
plt.axhline(color='black')
plt.axvline(color='black')
```

## Apply PCA

```python
from sklearn.decomposition import PCA
```

```python
p = PCA(n_components=2, random_state=101)
data_pca = p.fit_transform(data[0])
```

## Plot results of the PCA

```python
plt.scatter(x=data_pca[:,0], y=data_pca[:,1])
plt.axhline(color='black')
plt.axvline(color='black')
```
